```python
#Q3)a) A Vandermonde matrix is a matrix generated from a vector in which each column
#of the matrix is an integer power starting from zero. So, if I have a column vector [x1, x2, . . . , xN] T,
#then the associated (square) Vandermonde matrix would be 
#Use what you learned about broadcasting in the previous problem to write a
#function that will produce a Vandermonde matrix for a vector [1, 2, . . . ,N]T
#for any N. Do it without using a loop. Here’s a stub to get you started:
#Use your function for N = 12, store it in variable named vander, and print the result.

import numpy as np

def vandermonde(N):
    # Create the vector [1, 2, ..., N]
    vec = np.arange(N) + 1
    
    # Generate the Vandermonde matrix using broadcasting
    vander = vec ** np.arange(N).reshape(-1, 1)
    
    return vander

# Use the function for N = 12
vander = vandermonde(12)

# Print the resulting Vandermonde matrix
print("Vandermonde matrix for N=12:")
print(vander)
```

```python
#Q3)b) Now, let’s make a pretend linear system problem with this matrix. Create a vector
#of all ones, of length 12 and call it x. Perform a matrix-vector multiplication of
#vander with the vector you just created and store that in a new vector and call it b. Print the vector b.

import numpy as np

def vandermonde(N):
    vec = np.arange(N) + 1
    vander = vec ** np.arange(N).reshape(-1, 1)
    return vander

# Create the Vandermonde matrix for N = 12
vander = vandermonde(12)

# Create a vector of all ones of length 12
x = np.ones(12)

# Perform matrix-vector multiplication
b = vander @ x  # or you can use np.dot(vander, x)

# Print the resulting vector b
print("Vector b (result of matrix-vector multiplication):")
print(b)
```

```python
#Q3)c) First, solve the linear system the naïve way, pretending like you don’t know x.
#Import numpy.linalg, invert V and multiply it by b. Print out your result. What
#should you get for your answer? If the answer is different than what you expected, write a sentence about that difference.

import numpy as np
import numpy.linalg as la

def vandermonde(N):
    vec = np.arange(N) + 1
    vander = vec ** np.arange(N).reshape(-1, 1)
    return vander

# Create the Vandermonde matrix for N = 12
vander = vandermonde(12)

# Create a vector of all ones of length 12
x = np.ones(12)

# Perform matrix-vector multiplication
b = vander @ x  # This is our result vector

# Solve the linear system by inverting the Vandermonde matrix
x_hat = la.inv(vander) @ b

# Print the resulting vector x_hat
print("Result of solving the linear system (x_hat):")
print(x_hat)
```

```python
#Q3)d) Now, solve the same linear system using solve. Print out the result. Does it seem
#more or less in line with what you’d expect?

import numpy as np
import numpy.linalg as la

def vandermonde(N):
    vec = np.arange(N) + 1
    vander = vec ** np.arange(N).reshape(-1, 1)
    return vander

# Create the Vandermonde matrix for N = 12
vander = vandermonde(12)

# Create a vector of all ones of length 12
x = np.ones(12)

# Perform matrix-vector multiplication to create b
b = vander @ x  # This is our result vector

# Solve the linear system using np.linalg.solve
x_hat_solve = la.solve(vander, b)

# Print the resulting vector x_hat_solve
print("Result of solving the linear system using solve (x_hat_solve):")
print(x_hat_solve)
```
